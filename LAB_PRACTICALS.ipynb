```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
name = ['Vatsal','Palak','Priya','Riya','Mahi']
cgpa = [8,8,7,5,6]

plt.xlabel('Names')
plt.ylabel('Cgpa')

plt.bar(name, cgpa, color = ['red','pink','yellow','green','blue'])
```

```python
names = ['vatsal','palak','meera','shivam','mahir']
cgpa = [9.55,9.60,5.90,6.90,8.20]

plt.pie(x = cgpa, labels = names, autopct = "%0.2f%%")
```

```python
names = ['vatsal','palak','meera','shivam','mahir']
cgpa = [9.55,9.60,5.90,6.90,8.20]
explode = [0,0,0.1,0,0.1]
plt.pie(x = cgpa, labels = names, autopct = "%0.2f%%", explode = explode)
```

```python
price = [25000, 50000, 20000, 15000, 30000]
year = [2010, 2015, 2016, 2020, 2024]

plt.xlabel('Prices')
plt.ylabel('Year')

plt.scatter(price,year)
```

```python
profit = [2500, 3000, 3500, 4000, 4500]
year = [2010, 2015, 2016, 2020, 2024]

plt.xlabel('Profit')
plt.ylabel('Year')

plt.plot(profit,year)
```

```python
year = [2010, 2015, 2020, 2021, 2024]
profit = [2500, 3000, 3500, 4000, 4500]
sales = [20000, 10000, 25000, 20000, 50000]

plt.xlabel('year')
plt.ylabel('sales/profit')

plt.plot(year, profit, color='pink')
plt.plot(year, sales, color='blue')
```
