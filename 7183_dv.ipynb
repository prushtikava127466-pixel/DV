# LINEGRAPH

```python
profit = [2500, 3000, 3500, 4000, 4500]
year = [2010, 2015, 2016, 2020, 2024]

plt.xlabel('Profit')
plt.ylabel('Year')

plt.plot(profit,year)
```

```python
year = [2010, 2015, 2020, 2021, 2024]
profit = [2500, 3000, 3500, 4000, 4500]
sales = [20000, 10000, 25000, 20000, 50000]

plt.xlabel('year')
plt.ylabel('sales/profit')

plt.plot(year, profit, color='pink')
plt.plot(year, sales, color='blue')
```

# LEGEND

```python
year = [2010, 2012, 2014, 2016, 2018, 2020]
profit = [2500, 3000, 3500, 4000, 4500, 5000]
sales = [20000, 30000, 15000, 25000, 35000, 40000]

plt.xlabel('year')
plt.ylabel('sales/profit')

plt.plot(year, profit, color='pink', label = 'year vs profit')
plt.plot(year, sales, color='blue', label = 'year vs sales')

plt.legend()
```

# 3d scatterplot

```python
x = [1,4,7,10]
y = [2,5,8,11]
z = [3,6,9,12]

fig,axis = plt.subplots(figsize = (8,8),subplot_kw = {'projection': '3d'})

axis.scatter3D(x, y, z, color='pink')   
```
